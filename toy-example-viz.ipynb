```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 28
plt.rcParams['mathtext.fontset'] = 'stix'  # The setting of math font
```

```python
def f(x):
    return x ** 2


_, ax = plt.subplots(
    figsize=(10, 5)
)

LB, UB = -10, 10
dx = np.linspace(LB, UB, 10000)
rng = np.random.RandomState(42)

ax.set_ylim(-0.5, 25.5)
lines = [
    ax.plot(dx, f(dx), color="black", linestyle="dashed")[0],
    ax.plot(dx, np.minimum(f(dx), 1), color="red")[0],
    ax.fill_between([-1, 1], [-0.5, -0.5], [25.5, 25.5], alpha=0.2, hatch=r"\\"),
]
ax.grid()
labels = ["$f(x)$", "$\min(f(x), 1)$", "Local space"]
ax.legend(
    handles=lines,
    labels=labels,
    loc="upper center",
    fontsize=18,
    bbox_to_anchor=(0.5, -0.1),
    ncols=3,
)

plt.savefig("figs/toy-example-difference.png", bbox_inches="tight")
```

```python
def variance(L):
    return 4/3 * (2/5/L - 1/3/L**2)

L = np.linspace(1, 20, 1000)

plt.rcParams["font.size"] = 20
plt.plot(L, variance(L), color="blue", linestyle="dashed", label="Na√Øve")
plt.hlines(variance(1), 1, 20, color="red", label="Ours")
plt.xlabel("$L$", fontsize=20)
plt.ylabel("$\sigma^2$", fontsize=20)
plt.grid()
plt.legend(loc="upper right")
plt.savefig("figs/length-vs-variance.png", bbox_inches="tight")
```
